```python
import numpy as np

def calcular_potencia_ativa(V_i, V_j, delta_i, delta_j, theta, Z):
    # Cálculo da potência ativa usando a fórmula
    P_ij = (V_i**2 / Z) * np.cos(np.radians(theta)) - (V_i * V_j / Z) * np.cos(np.radians(delta_i)  - np.radians(delta_j) + np.radians(theta))
    return P_ij
    print(f'Potência ativa P_ij: {P_ij:.4f} pu')
    
# Definindo valores de exemplo
V_i = 1.05  # Tensão no nó i (em pu)
V_j = 1.00  # Tensão no nó j (em pu)
delta_i = np.radians(30)  # Ângulo de fase no nó i (em radianos)
delta_j = np.radians(15)  # Ângulo de fase no nó j (em radianos)
theta = np.radians(5)      # Ângulo de fase associado à impedância (em radianos)
Z = 0.1                     # Impedância da linha (em pu)

# Calculando a potência ativa
P_ij = calcular_potencia_ativa(V_i, V_j, delta_i, delta_j, theta, Z)

# Exibindo o resultado


a=1
```

```python
theta = range(0, 360, 5)  # Ângulo de fase associado à impedância (em graus)
thetas = []
results = []
for t in theta:
    P_ij = calcular_potencia_ativa(V_i, V_j, delta_i = 30, delta_j = 15, theta = t, Z = 0.1)
    #print(f'Potência ativa P_ij para theta = {t}°: {P_ij:.4f} pu')
    thetas.append(t)
    results.append(P_ij)

import matplotlib.pyplot as plt
# Plotando os resultados
plt.plot(thetas, results, marker='o')
plt.title('Potência Ativa P_ij em função de Theta para delta_i = 30° e delta_j = 15°')
plt.xlabel('Theta (graus)')
plt.ylabel('Potência Ativa P_ij (pu)')
plt.grid()
plt.axhline(0, color='gray', linestyle='--')
plt.show()
```
